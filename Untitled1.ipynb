```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
resnet34 = pd.read_csv('information/resnet34/ResNet34_9-5-2019-19-18-29_inf.csv')
resnext50 = pd.read_csv('information/resnext50/SEResNext50_11-5-2019-3-23-37_inf.csv')
senet150 = pd.read_csv('information/senet154/SENet154_10-5-2019-11-31-39_inf.csv')
```

```python
deeplab_1 = pd.read_csv('information/deeplab/DeepLab_14-5-2019-1-44-44_inf.csv')
deeplab_2 = pd.read_csv('information/deeplab/DeepLab_15-5-2019-1-59-6_inf.csv')
deeplab_3 = pd.read_csv('information/deeplab/DeepLab_15-5-2019-3-23-48_inf.csv')
deeplab_4 = pd.read_csv('information/deeplab/DeepLab_15-5-2019-10-50-28_inf.csv')
```

```python
deeplab = pd.DataFrame()
```

```python
deeplab = deeplab.append(deeplab_1).append(deeplab_2).append(deeplab_3).append(deeplab_4)
```

```python
deeplab.to_csv('deeplab_normal.csv', index=False)
```
